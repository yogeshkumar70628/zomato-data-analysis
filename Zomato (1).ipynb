# Location Analysis

```python
cities = df['city'].value_counts().head(10)
cities.plot(kind = 'bar',color = 'skyblue')
plt.title('Top 10 cities with most Resturants')
plt.xlabel('city')
plt.ylabel('Number of Resturants')
plt.show()
```

***Chennai leads with the highest number of restaurants, followed by Mumbai and Bangalore.***

***Tier-1 cities dominate the list, showing strong food culture and customer demand.***

***Smaller cities like Ajmer and Kochi also appear, suggesting Zomato’s reach is not limited to metros.***

***Business implication: High competition in Chennai/Mumbai/Bangalore, while smaller cities may offer growth opportunities with less saturation.***

```python
# Rating Distribution Across Cities
top = df['city'].value_counts().head(5).index
sns.boxplot(x='city', y='aggregate_rating', data=df[df['city'].isin(top)])
plt.xticks(rotation=45)
plt.title("Ratings Distribution by City")
plt.show()
```

***Big metros (Mumbai, Bangalore, Chennai) dominate in restaurant numbers but also show diverse quality.***

***Smaller cities (Lucknow, Pune) have fewer restaurants but more consistent ratings.***

***Across all cities, 4.5+ ratings are rare gems — valuable benchmarks for excellence.***


# Cuisine Analysis


```python
Cuisine_Analysis = df['Food_Category'].value_counts().head(7)
plt.figure(figsize = (12,5))
plt.bar(Cuisine_Analysis.index , Cuisine_Analysis.values,color = 'skyblue')
plt.title('Top 10 Cuisine')
plt.xlabel('Cuisine')
plt.ylabel('Number of Resturents')
plt.show()
```

***Fusion cuisines are emerging trends — offering multiple cuisines could attract broader audiences.***

***Niche categories (like Bakery) may provide opportunities for specialization and differentiation.***

```python
# Cuisine Variety vs Rating
plt.figure(figsize=(10,5))

plt.scatter(df['Food_Category'], df['aggregate_rating'], color='orange')

plt.title("Cuisine Variety vs Rating")
plt.xlabel("Number of Cuisines Offered")
plt.ylabel("Restaurant Rating")

plt.show()
```

Observation: Restaurants offering more cuisines don’t necessarily achieve higher ratings.

Trend: The scatter points show that ratings cluster around 3–4, regardless of how many cuisines are offered.

Key insight: Quality matters more than variety. Restaurants with fewer cuisines can still achieve high ratings if they focus on excellence.

Business implication: Expanding menus may attract customers, but maintaining consistency and quality is the real driver of ratings.

# Price Range and Rating

```python
price_rating = df.groupby('price_range')['aggregate_rating'].mean()

price_rating.plot(kind='bar')
plt.title("Price Range vs Average Rating")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.show()
```

***Insight: Average ratings increase with higher price ranges.***

***Business takeaway:***

***Premium restaurants (higher price ranges) tend to deliver better customer satisfaction.***

***Budget restaurants may need to improve quality to compete, while mid-range restaurants should highlight their balance of value and quality.***

```python
# Average Cost for Two by Price Category
sns.boxplot(x='price_range', y='average_cost_for_two', data=df)
plt.title("Cost for Two by Price Range")
plt.show()
```

***Insight: Higher price ranges show wider spreads and higher median costs for two people.***

***Business takeaway:***

***Customers expect higher costs in premium categories, but consistency matters — wide spreads suggest variability in perceived value.***

***Mid-range restaurants can position themselves as reliable and affordable, while premium restaurants must justify higher costs with superior service and experience.***

# ***Restaurant Features***

```python
features = df[['delivery','takeaway']].count()
plt.figure(figsize=(6,4))
plt.bar(features.index, features.values, color='pink')

plt.title("Restaurant Service Features")
plt.xlabel("Features")
plt.ylabel("Number of Restaurants")

plt.show()
```

***Insight: Almost all restaurants offer both delivery and takeaway (over 130k each).***

***Business takeaway:***

***These services are now industry standards — customers expect them by default.***

***Restaurants not offering delivery or takeaway risk losing relevance and reach.***

***Competitive edge lies not in offering these services, but in speed, reliability, and packaging quality.***

```python
# Delivery column counts
delivery_counts = df['delivery'].value_counts()

# Takeaway column counts
takeaway_counts = df['takeaway'].value_counts()

# Combine into one DataFrame
comparison = pd.DataFrame({
    'Delivery': delivery_counts,
    'Takeaway': takeaway_counts
})

print(comparison)
```

```python
comparison.plot(kind='bar', color=['skyblue','pink'])
plt.title("Delivery vs Takeaway Service Distribution")
plt.xlabel("Service Availability")
plt.ylabel("Number of Restaurants")
plt.show()
```

***Insight: Delivery and takeaway availability are nearly identical, with only a small difference in counts.***

***Business takeaway:***

***Restaurants should treat delivery and takeaway as complementary channels.***

***Optimizing both ensures maximum customer convenience.***

***Marketing should highlight seamless service experience across dine-in, delivery, and takeaway.***
